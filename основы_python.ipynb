# Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(s):
    """Возвращает количество гласных букв (a, e, i, o, u) в строке."""
    vowels = set('aeiouAEIOU')
    return sum(1 for ch in s if ch in vowels)

# Тесты
assert count_vowels("hello") == 2
assert count_vowels("rhythm") == 0
assert count_vowels("aeiou") == 5
assert count_vowels("") == 0
assert count_vowels("HELLO") == 2
print("Задача 1: все тесты пройдены")
print("count_vowels('hello') =", count_vowels("hello"))
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def all_unique(s):
    """True если каждый символ встречается ровно 1 раз, иначе False."""
    return len(s) == len(set(s))

# Тесты
assert all_unique("abcde") == True
assert all_unique("abcda") == False
assert all_unique("") == True
assert all_unique("a") == True
print("Задача 2: все тесты пройдены")
print("all_unique('abcde') =", all_unique("abcde"))
print("all_unique('abcda') =", all_unique("abcda"))
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_ones(n):
    """Количество единичных битов в двоичном представлении числа."""
    return bin(n).count('1')

# Тесты
assert count_ones(0) == 0
assert count_ones(1) == 1
assert count_ones(7) == 3    # 111 -> три единицы
assert count_ones(8) == 1    # 1000 -> одна единица
assert count_ones(255) == 8  # 11111111 -> восемь единиц
print("Задача 3: все тесты пройдены")
print("count_ones(255) =", count_ones(255), " (255 = 11111111 в двоичной)")
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    """Число шагов перемножения цифр до получения однозначного числа."""
    steps = 0
    while n >= 10:
        product = 1
        for d in str(n):
            product *= int(d)
        n = product
        steps += 1
    return steps

# Тесты
assert magic(39) == 3
assert magic(4) == 0
assert magic(999) == 4
assert magic(10) == 1   # 1*0 = 0
print("Задача 4: все тесты пройдены")
print("magic(39) =", magic(39))
print("magic(999) =", magic(999))
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import math

def mse(pred, true):
    """Среднеквадратическое отклонение (RMSE) между двумя векторами."""
    n = len(pred)
    return math.sqrt(sum((p - t) ** 2 for p, t in zip(pred, true)) / n)

# Тесты
assert abs(mse([1, 2, 3], [1, 2, 3])) < 1e-9       # одинаковые -> 0
assert abs(mse([0, 0], [1, 1]) - 1.0) < 1e-9
assert abs(mse([1, 2], [3, 4]) - 2.0) < 1e-9
print("Задача 5: все тесты пройдены")
print("mse([0,0], [1,1]) =", mse([0,0], [1,1]))
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида "(n1**p1)(n2**p2)…(nk**pk)" представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть "(2**5)(5)(7**2)(11)". Написать тесты для кода

```python
def prime_factors(n):
    """Разложение числа на простые множители вида (n1**p1)(n2**p2)..."""
    if n < 2:
        return ""
    factors = {}
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors[d] = factors.get(d, 0) + 1
            n //= d
        d += 1
    if n > 1:
        factors[n] = factors.get(n, 0) + 1
    result = ""
    for base in sorted(factors):
        exp = factors[base]
        result += f"({base})" if exp == 1 else f"({base}**{exp})"
    return result

# Тесты
assert prime_factors(86240) == "(2**5)(5)(7**2)(11)"
assert prime_factors(12) == "(2**2)(3)"
assert prime_factors(7) == "(7)"
assert prime_factors(1) == ""
print("Задача 6: все тесты пройдены")
print("prime_factors(86240) =", prime_factors(86240))
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку "It is impossible". Написать тесты для кода

```python
def pyramid(number):
    """Можно ли число представить как 1^2+2^2+...+k^2. Вернуть k или 'It is impossible'."""
    k = 1
    total = 0
    while total < number:
        total += k * k
        if total == number:
            return k
        k += 1
    return "It is impossible"

# Тесты
assert pyramid(1) == 1        # 1^2 = 1
assert pyramid(5) == 2        # 1+4 = 5
assert pyramid(14) == 3       # 1+4+9 = 14
assert pyramid(30) == 4       # 1+4+9+16 = 30
assert pyramid(3) == "It is impossible"
print("Задача 7: все тесты пройдены")
print("pyramid(14) =", pyramid(14))
print("pyramid(3) =", pyramid(3))
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(n):
    """Сбалансированное ли число: суммы цифр слева и справа от середины равны."""
    digits = [int(d) for d in str(n)]
    length = len(digits)
    if length <= 2:
        return True  # нет левой и правой части
    if length % 2 == 1:
        # нечётное: одна средняя цифра
        mid = length // 2
        left = digits[:mid]
        right = digits[mid + 1:]
    else:
        # чётное: две средних цифры
        mid1 = length // 2 - 1
        mid2 = length // 2
        left = digits[:mid1]
        right = digits[mid2 + 1:]
    return sum(left) == sum(right)

# Тесты
assert is_balanced(1) == True        # одна цифра
assert is_balanced(121) == True      # 1 == 1
assert is_balanced(123) == False     # 1 != 3
assert is_balanced(10001) == True    # [1,0] vs [0,1]
assert is_balanced(1221) == True     # [] vs [] (пустые части)
print("Задача 8: все тесты пройдены")
print("is_balanced(121) =", is_balanced(121))
print("is_balanced(10001) =", is_balanced(10001))
print("is_balanced(123) =", is_balanced(123))
```
